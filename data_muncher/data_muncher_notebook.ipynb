## Tratamento

```python
brute_df.head(20)
```

```python
# Id dos grupos
number_bq = 1
number_dq = 2
number_feature = 3
number_data = 4
```

```python
group_bq = brute_df[['id_business_question', 'business_question']].drop_duplicates()
group_bq['group'] = number_bq
group_bq.rename(columns={'id_business_question': 'id', 'business_question': 'label'}, inplace=True)
```

```python
group_bq
```

```python
group_data_question = brute_df[['id_data_question', 'data_question']].drop_duplicates()
group_data_question['group'] = number_dq
group_data_question.rename(columns={'id_data_question': 'id', 'data_question': 'label'}, inplace=True)
```

```python
group_data_question
```

```python
group_feature = brute_df[['id_feature', 'feature']].drop_duplicates()
group_feature['group'] = number_feature
group_feature.rename(columns={'id_feature': 'id', 'feature': 'label'}, inplace=True)
```

```python
group_feature
```

```python
group_data = brute_df[['id_dado', 'dado']].drop_duplicates()
group_data['group'] = number_data
group_data.rename(columns={'id_dado': 'id', 'dado': 'label'}, inplace=True)
```

```python
group_data
```

```python
df_nodes = pd.concat([group_bq, group_data_question, group_feature, group_data], ignore_index=True)
```

```python
df_nodes
```

```python
from_bq_to_dq = brute_df[['id_business_question', 'id_data_question']].drop_duplicates()
```

```python
from_bq_to_dq.rename(columns={'id_business_question': 'from', 'id_data_question': 'to'}, inplace=True)
```

```python
from_dq_to_ft = brute_df[['id_data_question', 'id_feature']].drop_duplicates()
```

```python
from_dq_to_ft.rename(columns={'id_data_question': 'from', 'id_feature': 'to'}, inplace=True)
```

```python
from_ft_to_data = brute_df[['id_feature', 'id_dado']].drop_duplicates()
```

```python
from_ft_to_data.rename(columns={'id_feature': 'from', 'id_dado': 'to'}, inplace=True)
```

```python
df_edges = pd.concat([from_bq_to_dq, from_dq_to_ft, from_ft_to_data])
```

```python
df_edges['from'] = df_edges['from'].astype(int)
df_edges['to'] = df_edges['to'].astype(int)

df_nodes['id'] = df_nodes['id'].astype(int)
```

Fiz isso para o pandas se virar convertendo os int64 em int, porque a lib json não consegue converter em json in64

```python
node_list = df_nodes.to_json(orient='records', force_ascii=False)
```

```python
edges_list = df_edges.to_json(orient='records', force_ascii=False)
```

```python
result = {'nodes': json.loads(node_list), 'edges': json.loads(edges_list)}
```

```python
result
```

```python
json_create(result)
```
